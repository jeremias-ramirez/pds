considerando la relacion de Euler con las senoidales, w sería el angulo y por lo tanto
puede ir desde 0 a 2*pi, las frecuencias positivas van desde 0 a pi.
Con la frecuncia de muestreo de 10khz entonces los puntos para analizar son, considerando 
solamente las frecuencias positivas, tengo una resolución frecuencial de df = 1/fm



```python
fm = 10e3
fre = np.linspace(0, math.pi, int(math.pi*fm))
numerador = 1
denominador = 1 - 0.5 * np.exp(1j*fre)**-1 + 0.25 * np.exp(1j*fre)**-2
h_z = numerador / denominador

plt.plot(fre, np.abs(h_z))
```
